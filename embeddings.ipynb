# Lib

```python
import boto3
import json
import os
```

```python
env_vars = !cat .env

for var in env_vars:
    key, value = var.split("=", 1)
    os.environ[key] = value.replace('"', "")


# Cria o cliente Bedrock
client = boto3.client(service_name='bedrock-runtime', region_name='us-east-1')
```

# playground

```python
roupa = "Calça jeans"


response = client.invoke_model(
    body=json.dumps({
        "inputText": roupa
    }),
    modelId='amazon.titan-embed-text-v1',
    contentType='application/json',
    accept='application/json',
)

response_body = json.loads(response['body'].read())

print(response_body['embedding'])
```

Os **embeddings** são representações matemáticas de dados que preservam suas relações semânticas. Em termos simples, eles convertem palavras, frases, imagens ou outros tipos de dados em vetores numéricos. Esses vetores capturam o significado dos dados, de forma que elementos com contextos ou sentidos semelhantes fiquem próximos uns dos outros no espaço vetorial.

O uso de embeddings é muito comum na inteligência artificial, pois possibilita que computadores lidem com dados complexos, como linguagem natural e imagens, de maneira mais eficiente e precisa.
