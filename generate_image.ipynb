# Playground

```python
client = boto3.client(
    'bedrock-runtime', 
    region_name='us-east-1',
    aws_access_key_id=os.environ['AWS_ACCESS_KEY_ID'],
    aws_secret_access_key=os.environ['AWS_SECRET_ACCESS_KEY'],
    aws_session_token=os.environ['AWS_SESSION_TOKEN']
)

def get_text_prompt():
    print("Entre com o prompt para gerar imagem (in English):")
    return input()

def get_file_name():
    print("Entre com o nome do arquivo para salvar a imagem (ex: image.png):")
    return input()

def get_image_config(prompt):
    return json.dumps({
        "taskType": "TEXT_IMAGE",
        "textToImageParams": {
            "text": prompt
        },
        "imageGenerationConfig": {
            "numberOfImages": 1,
            "height": 1024,  # Valores suportados: 1024, 768, 512
            "width": 1024,   # Valores suportados: 1024, 768, 512
            "cfgScale": 8.0,
            "seed": 0
        }
    })
```

```python
text_prompt = get_text_prompt()
file_name = get_file_name()

image_config = get_image_config(text_prompt)
response = client.invoke_model(
    modelId='amazon.nova-canvas-v1:0',
    contentType='application/json',
    accept='application/json',
    body=image_config
)

response_body = json.loads(response['body'].read())
base64_image = response_body['images'][0]
based64_bytes = base64.b64decode(base64_image)

with open(file_name, 'wb') as image_file:
    image_file.write(based64_bytes)

print(f'Imagem salva como {file_name}')
```

```python
client = boto3.client(
    'bedrock-runtime', 
    region_name='us-east-1',
    aws_access_key_id=os.environ['AWS_ACCESS_KEY_ID'],
    aws_secret_access_key=os.environ['AWS_SECRET_ACCESS_KEY'],
    aws_session_token=os.environ['AWS_SESSION_TOKEN']
)

def get_text_prompt():
    print("Entre com o prompt para gerar imagem (in English):")
    return input()

def get_file_name():
    print("Entre com o nome do arquivo para salvar a imagem (ex: image.png):")
    return input()

def get_image_config(prompt):
    return json.dumps({
        "taskType": "TEXT_IMAGE",
        "textToImageParams": {
            "text": prompt
        },
        "imageGenerationConfig": {
            "numberOfImages": 1,
            "height": 1024,  # Valores suportados: 1024, 768, 512
            "width": 1024,   # Valores suportados: 1024, 768, 512
            "cfgScale": 10, 
            "quality": "standard",
            "seed": 0,
            "numberOfImages": 1
        }
    })
```
